# **Feature Selection**
Selecting the most relevant features to improve model performance.

```python
df.drop("id",axis=1,inplace=True)
```

```python
df["diagnosis"]=[1 if row=="M" else 0 for row in df["diagnosis"] ]
```

```python
df.drop("Unnamed: 32",axis=1 , inplace=True)
```

# **Splitting the Dataset**
Dividing the data into training and testing sets to evaluate the model's performance.

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

# **Building the Logistic Regression Model**
Initializing and training the Logistic Regression model using the training dataset.

```python
x_train, x_test , y_train , y_test=train_test_split(x,y,random_state=42,test_size=.2)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_predic=model.predict(x_test)
```

```python
total= accuracy_score(y_test,y_predic)*100
total
```
